```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# frequência
x = np.arange(1,11)

# reatância indutiva
y = (270.1298699, 605.633802, 787.8787874, 951.6129029, 1214.285715, 1423.076924, 1702.127661, 1953.49000, 2000.00000, 2444.444445)
```

```python
L = 37.8
Xl = 2*np.pi*L*x
```

```python
# Legenda
plt.title("Comparação dos dados experimentais e cálculados")
plt.xlabel("frequência (kHz)")
plt.ylabel("Reatividade Indutiva(" + r'$\Omega$)')
plt.scatter(x, y, label="Dados experimentais", color="r")
plt.plot(x, Xl, label="Dados Calculados", color="k")
plt.legend()
plt.savefig("figura1.png", dpi=1200)
#plt.plot(x,y)
```
